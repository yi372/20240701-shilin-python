```python
import numpy as np
import pandas as pd

df = pd.DataFrame({'key1':['a', 'a', 'b', 'b', 'a'],
                  'key2':['one', 'two', 'one', 'two', 'one'],
                  'data1': np.random.randn(5),
                  'data2':np.random.randn(5)})
df
```

```python
df.T
```

```python
#因為key1,key2,data1,data2沒有重覆,可以stack()
df.stack()
```

```python
df1 = df.set_index(['key1','key2'])
df1
```

```python
#會出錯,因為值有重覆
#df1.unstack(level='key1')
#df1.unstack(level='key2')
```

```python
df2 = df.groupby(by=['key1','key2']).max()
df2
```

```python
df2.unstack(level='key2')
```
