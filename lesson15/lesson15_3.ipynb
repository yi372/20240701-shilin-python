```python
t = np.arange(0,5,0.1)
y1 = np.sin(t * np.pi)
y2 = np.cos(t * np.pi) 
import matplotlib.pyplot as plt
figure = plt.figure()
axes = figure.add_subplot()
axes.plot(t,y1)
axes.plot(t,y2)
plt.show()
```

```python
import matplotlib.pyplot as plt
labels = ['Nokia', 'Samsung', 'Apple', 'Lumia']
values = [20, 30, 45, 10]
explode = [0.1,0,0,0]
fig, ax = plt.subplots()
ax.pie(values, explode=explode, labels=labels, autopct='%1.1f%%',
       shadow=True, startangle=90,colors =['yellow', 'green', 'red', 'blue'])
plt.show()
```
