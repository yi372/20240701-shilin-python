```python
import random
with open("names.txt",encoding = "utf-8")as file:
    content = file.read()
names_lists = content.split()
names = random.choices(names_lists,k=50)
names
```

```python
import numpy as np
import pandas as pd
scores = np.random.randint(50,101,size = (50,5))
student_df = pd.DataFrame(scores,
             columns =['國文','英文','數學','地理','歷史'],
             index=names )
student_df
```

```python
#選取dataframe的一個欄位
display(student_df['國文'])
display(student_df['英文'])
```

```python
#選取dataframe的多個欄位
student_df[['國文','英文']]
```

```python
sum_series = student_df['國文'] + \
student_df['英文'] + \
student_df['數學'] + \
student_df['地理'] + \
student_df['歷史'] 
mean_series = sum_series / 5
mean_series
```

```python
student_df['總分'] = sum_series
student_df['平均'] = mean_series
student_df
```
