```python
import matplotlib.pyplot as plt
labels = ['Nokia', 'Samsung', 'Apple', 'Lumia']
values = [20, 30, 45, 10]
explode = [0.3,0,0,0]
fig, ax = plt.subplots()
ax.pie(values,
        explode=explode,
        labels=labels, 
        autopct='%1.1f%%',
        shadow=True,
        startangle=180,colors =['yellow', 'green', 'red', 'blue'])
plt.show()
```
